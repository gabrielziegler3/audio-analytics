```python
import scipy
import numpy as np
import matplotlib.pylab as plt

from scipy.signal import spectrogram
from IPython.display import Audio

plt.style.use('bmh')
%matplotlib inline
```

# Sample Audio Signal

```python
phase = 0
frame_size = 2048
sample_rate = 22050
freq = 440
t = np.arange(frame_size) / sample_rate
samples = np.sin(2 * np.pi * freq * t + phase)
time_vector = np.arange(frame_size) / sample_rate

_ = plt.figure(figsize=(14, 3))
_ = plt.plot(time_vector, samples)
```

```python
Audio(samples, rate=sample_rate, autoplay=True)
```

# Spectrogram

```python
_ = plt.figure(figsize=(10, 8))
_ = plt.specgram(samples, Fs=sample_rate)
```
